## Import libraries BeautifulSoup,urllib, nltk, download punkt

```python
from bs4 import BeautifulSoup
import urllib.request
import nltk
nltk.download('punkt')
```

## Get the response from the webpage http://php.net/

```python
response = urllib.request.urlopen('http://php.net/')
html = response.read()
print(html)
```

## Using BeautifulSoup and ntk tokenized HTML response

```python
soup = BeautifulSoup(html,"html5lib")
text_tokens = nltk.tokenize.word_tokenize(soup.get_text())
print(text_tokens)
```

## Download and import stopwords from nltk

```python
nltk.download('stopwords')
from nltk.corpus import stopwords
```

## Process tokens and if it i snot a stopword, caclulate count and put into a dictionary

```python
token_count_map ={}
for token in text_tokens:
    if token not in stopwords.words('english'): 
        if token in token_count_map.keys():
            count = token_count_map.get(token)
            count += 1
            token_count_map[token] = count        
        else:
            token_count_map[token] = 1         
    
```

## Print each token and its occurrence count

```python
for token,count in token_count_map.items():
    print (str(token) + ':' + str(count))  
```
